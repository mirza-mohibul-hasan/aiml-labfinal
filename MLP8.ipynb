```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load the data
data = np.array([
    [1, 1, 0, -1],
    [2, 0, 1, -1],
    [3, 0, -1, -1],
    [4, -1, 0, 1],
    [5, 0, 2, 1],
    [6, 0, -2, 1],
    [7, -2, 0, 1]
])
```

```python
# Extract features and labels
X = data[:, 1:3]
y = data[:, 3]
```

```python
# Create a scatter plot of the original 2D data
plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], marker='o', label='Class -1')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], marker='x', label='Class 1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Scatter Plot of Original 2D Data')
plt.legend()
plt.show()
```

# Gaussian kernel function and transform the data into 3D

```python
# Define the Gaussian kernel function
def gaussian_kernel(x1, x2, sigma=1.0):
    return np.exp(-np.sum((x1 - x2) ** 2) / (2 * sigma ** 2))
```

```python
# Apply the Gaussian kernel function to transform the data into 3D
Z = np.zeros((len(X), 2))  # Transformed 3D data
```

```python
for i in range(len(X)):
    Z[i, 0] = gaussian_kernel(X[i], np.array([1, 0]))  # Using ùõ∑1(ùë•1, ùë•2) = ùë•2^2 - 2x1 + 3
    Z[i, 1] = gaussian_kernel(X[i], np.array([0, 1]))  # Using ùõ∑2(ùë•1, ùë•2) = ùë•1^2 ‚àí 2ùë•2 + 3
```

```python
# Create a 3D scatter plot of the transformed data
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(Z[y == -1][:, 0], Z[y == -1][:, 1], y[y == -1], marker='o', label='Class -1')
ax.scatter(Z[y == 1][:, 0], Z[y == 1][:, 1], y[y == 1], marker='x', label='Class 1')
ax.set_xlabel('ùõ∑1(x1, x2)')
ax.set_ylabel('ùõ∑2(x1, x2)')
ax.set_zlabel('y')
ax.set_title('3D Scatter Plot of Transformed Data')
ax.legend()
plt.show()
```
